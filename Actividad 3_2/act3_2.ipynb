```python
# Reimportar librerías necesarias después del reinicio del estado
import pandas as pd
```

```python
# Cargar el archivo CSV nuevamente
file_path = "sandiego_estadosunidos.csv"
df = pd.read_csv(file_path)
```

```python
# Verificar las primeras filas del dataset
df.head()
```

```python
#Se identifican las columnas numéricas y en cada una se calcula la mediana, reemplazando los valores nulos.
numeric_cols = df.select_dtypes(include=['int', 'float']).columns
for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)
df.to_csv("output_median.csv", index=False)
print(df.isnull().sum())
```

```python
#Se identifican las columnas de tipo objeto, fecha y categoría, luego se aplican los valores anteriores para llenar los nulos.
object_cols = df.select_dtypes(include=['object', 'datetime', 'category']).columns
for col in object_cols:
    df[col].fillna(method='ffill', inplace=True)
df.to_csv("output_ffill.csv", index=False)
print(df.isnull().sum())
```

```python
#Se identifican las columnas de tipo objeto, fecha y categoría, luego se aplican los valores siguientes para llenar los nulos.
for col in object_cols:
    df[col].fillna(method='bfill', inplace=True)
df.to_csv("output_bfill.csv", index=False)
print(df.isnull().sum())
```

```python
#Se identifican las columnas de tipo objeto, fecha y categoría, luego se reemplazan los valores nulos con una cadena específica.
replacement_string = "Valor por defecto"
for col in object_cols:
    df[col].fillna(replacement_string, inplace=True)
df.to_csv("output_string.csv", index=False)
print(df.isnull().sum())
```

```python
#Se identifican las columnas numéricas y en cada una se reemplazan los valores nulos con una constante definida.
constant_value = 0
for col in numeric_cols:
    df[col].fillna(constant_value, inplace=True)
df.to_csv("output_constant.csv", index=False) 
print(df.isnull().sum())
```
